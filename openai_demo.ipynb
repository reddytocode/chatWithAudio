# Chat completition

```python

def ask_chat_gpt(message, callback):
    stream = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": message,
            }
        ],
        model="gpt-3.5-turbo",
        stream=True
    )
    cache = ""
    for chunk in stream:
        if a := chunk.choices[0].delta.content:
            print(a)
            cache += a
            if a in ".,;":
                callback(cache)
                cache = ""
    if cache:
        callback(cache)
            
```
